# Data Preparation
Two datasets were used for this project: the dogs_vs_cats dataset from week five of the class and a landscapes dataset from Kaggle.  Included in the zip file are 80 training images, 10 validation, and 4 to 5 testing images for each.  You may run the “code” notebook to get started.  However, I trained my models using 1000 training images, and if you wish to use the same dataset, then please follow these steps:
- Create a folder called ‘data’ in the same directory as the notebook files.
- Download the dog_vs_cats.zip file from Canvas.
  - https://canvas.harvard.edu/courses/164142/pages/week-5-2?module_item_id=1976712
- In the dogs_vs_cats.zip file, extract train.zip
- In the train.zip file, extract the train folder and put it in the ‘data’ folder.
- Download the landscapes dataset from Kaggle.
  - https://www.kaggle.com/datasets/theblackmamba31/landscape-image-colorization/data
- Extract the ‘landscape Images’ folder and put it in the ‘data’ folder.
- Run all cells in this notebook
  - This will create two new folders (‘dog_images’ and ‘landscape_images’) in the main directory.  These contain the training and validation subsets.


```python
# Libraries to run
import os, shutil
```

# Dogs

```python
# The path to the directory where the original dataset was uncompressed
dogs_dataset_dir = os.path.join('data', 'train')

# The directory where we will store our smaller dataset
dogs_base_dir = 'dog_images'

# Directories for our training, validation and test splits
dogs_train_dir = os.path.join(dogs_base_dir, 'train')
dogs_validation_dir = os.path.join(dogs_base_dir, 'validation')
dogs_test_dir = os.path.join(dogs_base_dir, 'test')
```

```python
if not os.path.exists(dogs_base_dir):
    os.mkdir(dogs_base_dir)

if not os.path.exists(dogs_train_dir):
    os.mkdir(dogs_train_dir)

if not os.path.exists(dogs_validation_dir):
    os.mkdir(dogs_validation_dir)

if not os.path.exists(dogs_test_dir):
    os.mkdir(dogs_test_dir)

if not os.path.exists(dogs_train_dir + '/dogs'):
    os.mkdir(dogs_train_dir + '/dogs')

if not os.path.exists(dogs_validation_dir + '/dogs'):
    os.mkdir(dogs_validation_dir + '/dogs')

if not os.path.exists(dogs_test_dir + '/dogs'):
    os.mkdir(dogs_test_dir + '/dogs')
```

```python
# Copy first 1000 dog images to train_dogs_dir
fnames = ['dog.{}.jpg'.format(i) for i in range(1000)]
for fname in fnames:
    src = os.path.join(dogs_dataset_dir, fname)
    dst = os.path.join(dogs_train_dir + '/dogs', fname)
    shutil.copyfile(src, dst)

# Copy next 500 dog images to validation_dogs_dir
fnames = ['dog.{}.jpg'.format(i) for i in range(1000, 1500)]
for fname in fnames:
    src = os.path.join(dogs_dataset_dir, fname)
    dst = os.path.join(dogs_validation_dir + '/dogs', fname)
    shutil.copyfile(src, dst)

# Copy next 500 dog images to test_dogs_dir
fnames = ['dog.{}.jpg'.format(i) for i in range(1500, 2000)]
for fname in fnames:
    src = os.path.join(dogs_dataset_dir, fname)
    dst = os.path.join(dogs_test_dir + '/dogs', fname)
    shutil.copyfile(src, dst)
```
